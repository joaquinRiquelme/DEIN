```python
# Librerias
import networkx as nx
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
```

```python
# se crea una matriz con valores aleatorios entre 0 y 1
# la matriz de adyacencia debe ser cuadrada
row = 100 ; cols = row
RandMatrix = np.random.randint(0,2, size = (row,cols))
```

```python
# Grafo y grafico a partir de la matriz de adyacencia
RandMatrix
G0 = nx.from_numpy_array(RandMatrix)
plt.figure(figsize=(10,10))
nx.draw(G0, with_labels=True, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
plt.title("Network from a Random Adjacency Matrix")
plt.show()
```

```python
# Grafo y grafico en base a nx.DiGraph() para dar direccion aleatoria
G = nx.from_numpy_array(RandMatrix, create_using=nx.DiGraph())
plt.figure(figsize=(10,10))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
plt.title("Network from a Random Adjacency Matrix")
plt.show()

```

```python
print(f"Nodes: {G.nodes()}")
print(f"Edges: {G.edges()}")
print(G)
```

```python
# densidad
density=nx.density(G)
print(density)
```

```python
num_node=G.number_of_nodes()
num_edges=G.number_of_edges()
print(f"Numero de nodos: {num_node}")
print(f"Numero de edges: {num_edges}")
print(row*(row-1)*density)
```

```python
def graph_feature(G):
    num_node=G.number_of_nodes()
    num_edges=G.number_of_edges()
    print(f"Numero de nodos: {num_node}")
    print(f"Numero de edges: {num_edges}")
    density=nx.density(G)
    print(f"Densidad: {density}")
```

```python
def graph_feature_2(X):
    # se crea una matriz con valores aleatorios entre 0 y 1
    # # la matriz de adyacencia debe ser cuadrada
    row = X ; cols = row
    RandMatrix = np.random.randint(0,2, size = (row,cols))
    
    # Grafo y grafico en base a nx.DiGraph() para dar direccion aleatoria
    G = nx.from_numpy_array(RandMatrix, create_using=nx.DiGraph())
    plt.figure(figsize=(10,10))
    nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
    plt.title("Network from a Random Adjacency Matrix")
    plt.show()
    
    num_node=G.number_of_nodes()
    num_edges=G.number_of_edges()
    print(f"Numero de nodos: {num_node}")
    print(f"Numero de edges: {num_edges}")
    density=nx.density(G)
    print(f"Densidad: {density}")
```

```python
# Eigen vector centrality: mide la influencia de un nodo en una red
eigenV_centrality=nx.eigenvector_centrality(G0)
print(eigenV_centrality)

eigenV_list=list(eigenV_centrality.values())
plt.plot(eigenV_list)
plt.show()
```

```python
min_val=min(eigenV_list)
max_val=max(eigenV_list)
print(f"Minimo: {min_val}", f"Maximo: {max_val}")

```
