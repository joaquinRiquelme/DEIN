```python
# Librerias
import networkx as nx
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
```

```python
import time
time2=700
col=100
row=col
max_val_time=[]
max_val_key=[]

start_time=time.time()
for i in range(time2):
    ranMatrix=np.random.randint(0,2,size=(row,col))
    G = nx.from_numpy_array(ranMatrix)
    eigenvector_centrality=nx.eigenvector_centrality(G)
    max_val=max(eigenvector_centrality.values())
    max_val_time.append(max_val)
    max_key=max(eigenvector_centrality, key=eigenvector_centrality.get)
    max_val_key.append(max_key)
    
end_time=time.time()
procces_time=end_time-start_time
print(f"Execution time: {procces_time:.6f} seconds")
```

```python
print(max_val_key)
print(max_val_time)
plt.plot(max_val_time)
```
