```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Crear un grafo
G = nx.Graph()
print(G)
```

```python
# a√±adir nodos
G.add_node("A")
print(G)
```

```python
# para mostrar o graficar la red, debemos utilizar lo siguiente
plt.figure(figsize=(3,3))
nx.draw(G, with_labels=True, node_color= 'blue', node_size=400, font_size=8, font_weight='bold' )
plt.show()
```

```python
G.add_nodes_from(["B","C","D","E","F"])
print(G)
plt.figure(figsize=(3,3))
nx.draw(G, with_labels=True, node_color= 'blue', node_size=400, font_size=8, font_weight='bold' )
plt.show()
```

```python
G.add_edges_from([("A","B"),("B","C"),("E","F"),("D","A"),("F","A")])
plt.figure(figsize=(3,3))
nx.draw(G, with_labels=True, node_color= 'blue', node_size=400, font_size=12, font_weight='bold' )
plt.show()
```

```python
M_adj = nx.adjacency_matrix(G)
print(M_adj)
M_adj

adjM = M_adj.todense()
print(adjM)
```

Grafo desde una matriz de adyacencia

```python
F=nx.from_numpy_array(adjM)
print(F)
```

```python
plt.figure(figsize=(3,3))
nx.draw(F, with_labels=True, node_color= 'green', node_size=400, font_size=12, font_weight='bold' )
plt.title("Network")
plt.show()
```

```python
adjM2 = adjM.copy()
adjM2[0,0]=1
F=nx.from_numpy_array(adjM2, create_using=nx.DiGraph())
plt.figure(figsize=(5,5))
nx.draw(F, with_labels=True, node_color= 'green', node_size=400, font_size=12, font_weight='bold' )
plt.title("Network")
plt.show()
```

```python
# tarea crear una matriz de 5x5 y que al menos 2 tenga una doble conexion entre si

m1 = np.matrix(np.ones(25))
m1.shape = (5,5)
print(m1)
m1[(0,0)]=0
m1[(1,1)]=0
m1[(2,2)]=0
m1[(3,3)]=0
m1[(4,4)]=0


G1=nx.from_numpy_array(m1, create_using=nx.DiGraph())
print(G1)

G2 = nx.relabel_nodes( G= G1, mapping={0: "a", 1: "b", 2: "c", 3: "d", 4:"e"})

plt.figure(figsize=(5,5))
nx.draw(G2, with_labels=True, node_color= 'yellow', node_size=800, font_size=12, font_weight='bold' )
plt.title("Network")
plt.show()

```
