```python
# Librerias
import networkx as nx
from networkx.algorithms import bipartite
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import time
```

```python
B=nx.Graph()
nodos_A=[3,2,1]
B.add_nodes_from(nodos_A,bipartite=0)
nodos_B=['c','b','a']
B.add_nodes_from(nodos_B,bipartite=1)
B.add_edge(1,'a')
B.add_edge(1,'b')
B.add_edge(2,'c')
B.add_edge(3,'c')
B.add_edges_from([(1,'a'),(1,'b'),(2,'c'),(3,'c')])


print(B)
nx.draw(B, with_labels=True, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
plt.title("Network from a Random Adjacency Matrix")
plt.show()
```

```python

pos = {}
pos.update((node, (1, i)) for i, node in enumerate(nodos_A))
pos.update((node, (2, i)) for i, node in enumerate(nodos_B))


nx.draw_networkx_nodes(B, pos, nodelist=nodos_A, node_color='lightblue', label='Conjunto A')
nx.draw_networkx_nodes(B, pos, nodelist=nodos_B, node_color='lightgreen', label='Conjunto B')
nx.draw_networkx_edges(B, pos, edge_color='gray')
nx.draw_networkx_labels(B, pos)

plt.title("Grafo Bipartito")
plt.show()
```
